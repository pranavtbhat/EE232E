```r
library(igraph)
library(MASS)
library(ggplot2)
library(irlba)

plot_core = function(g, core) {
    vertex_size = rep(5, vcount(g))
    vertex_color = rep("blue", vcount(g))
    
    if(core > 0){
        vertex_color[g$names == core] = "red"
        vertex_size[g$names == core] = 10  
    }
    
    plot(g, vertex.size = vertex_size, vertex.color = vertex_color, vertex.label=NA)
}

plot_comm = function(g, f, core) {
    vertex_size = rep(5, vcount(g))
    vertex_color = rep("blue", vcount(g))
    
    if(core > 0){
        vertex_color[g$names == core] = "red"
        vertex_size[g$names == core] = 10  
    }
    
    comm = f(g)
    cat("Modularity:", modularity(comm), "\nCommunity Sizes", sizes(comm), "\n")
    plot(g, vertex.size = vertex_size , vertex.label = NA , vertex.color = comm$membership, layout=layout.fruchterman.reingold)
}

options(repr.plot.width=4, repr.plot.height=4)
colors = c("red", "yellow", "green", "violet", "orange", "blue", "pink", "cyan") 
```

```r
g = read_graph("data/facebook_combined.txt", format="edgelist", directed=FALSE)
core_vertexes = which(degree(g) > 200)
```

```r
make_personal_network = function(g, u){
    nbors = sort(unlist(neighborhood(g, 1, u)))
    sg = induced_subgraph(g, nbors)
    sg$names = nbors
    
    return(sg)
}

mutual_friends = function(g, u, v){
    nu = neighbors(g, u)
    nv = neighbors(g, v)
    
    return(intersect(nu, nv))
}

embeddedness = function(g, core, v){
    return(length(mutual_friends(g, core, v)))
}

dispersion = function(g, core, node, W){
    disp = 0
    
    mfriends = mutual_friends(g, core, node)
    mg = delete_vertices(g, c(core, node))
    
    if(length(mfriends) >= 2){
        combs = combn(mfriends, 2)
        for(i in 1 : ncol(combs)){
            u = combs[1,i]
            v = combs[2,i]

            disp = disp + W[u,v]
        }
    }
    else{
        disp = 0
    }
    return(disp)
}
```

###  Plot the distribution of embeddedness ...

```r
emb_vec = c()

for(core in core_vertexes){
    sg = make_personal_network(g, core)
    v_core = which(sg$names == core)
    
    for(v in V(sg)){
        if(v == v_core)
            next
        emb_vec = c(emb_vec, embeddedness(sg, v_core, v))
    }
}
```

```r
hist(
    emb_vec, 
    breaks = 50, 
    main = "Embeddedness Distribution",
    xlab = "Embeddedness", 
    col="green"
)
```

### ... and dispersion over all personal networks created with core nodes from part 3.

```r
disp_vec = c()
for(core in core_vertexes){
    sg = make_personal_network(g, core)
    v_core = which(sg$names == core)
    W = distances(sg)
    
    for(v in V(sg)){
        if(v == v_core)
            next
        disp_vec = c(disp_vec, dispersion(sg, v_core, v, W))
    }
}
```

```r
hist(
    disp_vec, 
    breaks = 50, 
    main = "Dispersion Distribution",
    xlab = "Dispersion", 
    col="green"
)
```

### Plot 3 personal networks showing their community structure with colors and highlight the node with maximum dispersion, embeddness and the ratio

```r
plot_highlighted = function(g, core, v, text){
    comm = fastgreedy.community(g)
    vertex_size = rep(5, vcount(g))
    vertex_color = comm$membership
    edge_color = rep("lightgray", ecount(g))
    edge_sizes = rep(0.5, ecount(g))
    
    # Highlight  vertex
    vertex_color[v] = "red"
    vertex_size[v] = 30
    
    # Highlight edges
    elist = get.edge.ids(g, c(t(ends(g, E(g)[from(v)]))))
    edge_color[elist] = "red"
    edge_sizes[elist] = 1.0
    
    # 
    plot(
        g, 
        vertex.size = vertex_size, 
        vertex.label = NA, 
        vertex.color = vertex_color,
        edge.color=edge_color,
        edge.width=edge_sizes,
        layout=layout.fruchterman.reingold,
        main=text
    )
}
core_sample = sample(core_vertexes, 3)
```

### First Sample

```r
core = core_sample[1]

sg = make_personal_network(g, core)
v_core = which(sg$names == core)
W = distances(sg)

emb_vec = rep(0, vcount(sg))
disp_vec = rep(0, vcount(sg))

for(v in V(sg)){
    if(v == v_core){
        next
    }
    else{
        emb_vec[v] = embeddedness(sg, v_core, v)
    }
}

disp_vec[mapply(is.infinite, disp_vec)] = 0
```

```r
plot_highlighted(sg, v_core, which.max(emb_vec), "Max Embeddedness Highlighted")
```

```r
plot_highlighted(sg, v_core, which.max(disp_vec), "Max Dispersion Highlighted")
```

```r
plot_highlighted(sg, v_core, which.max(emb_vec/disp_vec), "Max E/D Highlighted")
```

```r
library(igraph)
library(MASS)
library(ggplot2)
library(irlba)
options(repr.plot.width=4, repr.plot.height=4)


g = read_graph("data/facebook_combined.txt", format="edgelist", directed=FALSE)
core_vertexes = which(degree(g) > 200)
core_sample = sample(core_vertexes, 10)
```

```r
get_community_features = function(g, u){
    sg = induced_subgraph(g, unlist(neighborhood(g, 1,  u)))
        
    comm_obj = fastgreedy.community(sg)
    comms = NULL
    
    for(i in 1 : length(comm_obj)){
        vs = unlist(comm_obj[i])
        
        if(length(vs) > 10){
            cg = induced_subgraph(sg, vs)
            
            ccoeff = transitivity(cg)
            mindex = modularity(fastgreedy.community(cg))
            density = edge_density(cg)
            comm_size = length(vs)
            dmeter = diameter(cg)

            comms = rbind(comms, c(ccoeff, mindex, density, comm_size, dmeter))
        }
    }
    
    return(comms)
}
```

```r
fvec = NULL

for(core in core_vertexes){
    fvec = rbind(fvec, get_community_features(g, 1))
}
```

```r
colnames(fvec) = c("Clustering Coefficient", "Modularity Index", "Density", "Size", "Diameter")
```

```r
# 5 gives best values
# Pick the two clusters with 160 each and call them communities. 
# So our desired communities are 2, 3
cl = kmeans(fvec, 2)
```

```r
cl
```

```r
assign_cluster <- function(x, centers) {
  # compute squared euclidean distance from each sample to each cluster center
  tmp <- sapply(seq_len(nrow(x)),
                function(i) apply(centers, 1,
                                  function(v) sum((x[i, ]-v)^2)))
  max.col(-t(tmp))  # find index of min distance
}
```

```r
for(core in core_vertexes){
    vvec = get_community_features(g, core)
    cat(core,"\t", assign_cluster(vvec, cl$centers), "\n")
}
```

```r
file_names = list.files("data/gplus/")
file_ids = sub("^([^.]*).*", "\\1", file_names)
ego_node_ids = unique(file_ids) 

cat("There are", length(ego_node_ids), "ego nodes")

ids_circles = numeric()
for (id in ego_node_ids) {
    circles_file = paste("data/gplus/" , id , ".circles" , sep="")
    circles_connect = file(circles_file , open="r")
    circles_content = readLines(circles_connect)
    close(circles_connect)

    if(length(circles_content) > 2){
        ids_circles = c(ids_circles, id)
    }
}
```

```r
cat("Total ego networks with > 2 circles = ", length(ids_circles))
```

```r
show_similarity = function(id){
    edges_file = paste("data/gplus/" , id  , ".edges" , sep="")
    circles_file = paste("data/gplus/" , id , ".circles" , sep="")
    circles_connect = file(circles_file , open="r")
    circles_content = readLines(circles_connect)
    
    circles = list()

    for (i in 1:length(circles_content)) {
        circle_nodes = strsplit(circles_content[i],"\t")
        circles = c(circles, list(circle_nodes[[1]][-1]))
    }

    close(circles_connect)

    g_network = read.graph(edges_file , format = "ncol" , directed=TRUE)  
    g_network = add.vertices(g_network, nv = 1, name = id)
    ego_node_index = which(V(g_network)$name==id) 

    add_edge_list = c()
    for (vertex in 1:(vcount(g_network) - 1)) {
        add_edge_list = c(add_edge_list, c(ego_node_index, vertex))
    }

    g_network = add_edges(g_network, add_edge_list)

    walktrap_comm = walktrap.community(g_network)
    infomap_comm = infomap.community(g_network)

    percentage = vector()
    percentage_circle = vector()
    
    for(m in 1:max(walktrap_comm$membership)){
        community_nodes = V(g_network)$name[which(walktrap_comm$membership == m)]

        for (n in 1:length(circles)) {
            common_nodes = intersect(community_nodes, circles[[n]])
            percent_circle = length(common_nodes)/length(circles[[n]])
            percentage_circle = c(percentage_circle, percent_circle)
        }
    }

    for(m in 1:max(infomap_comm$membership)){
        community_nodes = V(g_network)$name[which(infomap_comm$membership == m)]

        for (n in 1:length(circles)) {
            common_nodes = intersect(community_nodes, circles[[n]])
            percent = length(common_nodes)/length(circles[[n]])
            percentage = c(percentage_circle, percent_circle)
        }
    }

    cat("Number of Circles: ", length(circles))

    c_percentage = matrix(percentage_circle, nrow = max(walktrap_comm$membership), ncol = length(circles))
    colnames(c_percentage) = paste("GC ", 1 : length(circles), sep="")
    rownames(c_percentage) = paste("C ", 1 : max(walktrap_comm$membership), sep="")
    corrplot(c_percentage, method="circle", col="black", tl.cex = 2, tl.col = 'black', 
             main="Correlation for Fast Greedy")

    i_percentage = matrix(percentage, nrow = max(infomap_comm$membership), ncol = length(circles))
    colnames(i_percentage) = paste("GC ",    1:length(circles),     sep="")
    rownames(i_percentage) = paste("C ",    1:max(infomap_comm$membership), sep="")
    corrplot(i_percentage, method="circle", col="black", tl.cex = 2, tl.col = 'black',
            main="Correlation for Infomap")

}
```

```r
show_similarity(ids_circles[2])
```
